```python
eval_outputs = "/ytech_m2v5_hdd/workspace/kling_mm/yangsihan05/proj_text_enc/attn_pool_contrastive/eval_output_new"

import os
import json

layers = list(range(1, 38))
print(layers)
results_dict = {}
for layer in layers:
    # internvl3_5_last{layer}_2layers_final.json
    file_path = os.path.join(
        eval_outputs, f"ovis2_5_last{layer}_2layers_final.json"
    )
    with open(file_path, "r") as f:
        result = json.load(f)
        results_dict[37 - layer] = result["summary"]["accuracy_statistics"]["overall_accuracy"]
        # if layer == 1 or layer == 2:
        #     results_dict[37 - layer] = 0.5555 if layer == 1 else 0.5600
        print(f"Layer {37 - layer}: {results_dict[37 - layer]}")
    
# draw 折线图
import matplotlib.pyplot as plt
plt.figure(figsize=(14, 6))
plt.plot(list(results_dict.keys()), list(results_dict.values()), marker='o')
plt.title('Overall Accuracy vs Layer')
plt.xlabel('Layer')
plt.ylabel('Overall Accuracy')
plt.xticks(layers)
plt.grid()
plt.savefig('overall_accuracy_vs_layer.png', dpi=400)
plt.show()

```

```python
import json
import os

with open(intern35_path, "r") as f:
    intern35_result = json.load(f)

with open(intern35_ours_path, "r") as f:
    intern35_ours_result = json.load(f)
```

```python
intern35_result["results"]
```

```python
type_ours_bad = {}
type_ours_good = {}

for res_intern35, res_intern35_o in zip( intern35_result["results"], intern35_ours_result["results"] ):
    if res_intern35['is_correct'] != res_intern35_o['is_correct']:
        if res_intern35['is_correct']:
            type_q = f"{res_intern35['caption_type']}_{res_intern35['question_type']}"
            type_ours_bad[type_q] = type_ours_bad.get(type_q, 0) + 1
            print(f"Index: {res_intern35['index']}, InternVL3-5 is_correct: {res_intern35['is_correct']}, InternVL3-5-Ours is_correct: {res_intern35_o['is_correct']}")
            print(f"  Scores: InternVL3-5: {res_intern35['all_scores']}, InternVL3-5-Ours: {res_intern35_o['all_scores']}")
        else:
            type_q = f"{res_intern35['caption_type']}_{res_intern35['question_type']}"
            type_ours_good[type_q] = type_ours_good.get(type_q, 0) + 1
            print(f"Index: {res_intern35['index']}, InternVL3-5 is_correct: {res_intern35['is_correct']}, InternVL3-5-Ours is_correct: {res_intern35_o['is_correct']}")
            print(f"  Scores: InternVL3-5: {res_intern35['all_scores']}, InternVL3-5-Ours: {res_intern35_o['all_scores']}")
```
